#  Creating a data to view and fit in the regression model


```python

import numpy as np
import matplotlib.pyplot as plt


#  Creating features ( independent variable)
X = np.array([-7.0, - 4., -1., 2., 5. , 8., 11., 14. ])

#  Creating labels ( dependent variable)
Y = np.array([3., 6., 9., 12., 15., 18., 21. , 24.  ])

#  Visuakizing created data
plt.scatter(X, Y)
```

```python
#  Relationship of x and y ( independent and dependent variable)

Y == X + 10
```

#  Input & Output shapes for regression

```python
#  predicting house price

#  Inputs = no. of bedrooms, baths, cars ( 3)
#  Outputs = price of house ( 1)

house_param = tf.constant(["bedroom", "bath", "cars"])
house_price = tf.constant([939700])

house_param, house_price
```

```python
input_shape = X.shape
output_shape = Y.shape

input_shape, output_shape
```

```python
X[0], Y[0]
```

```python
#  converting nparrays to tensors ( default stype will be float32)

x = tf.constant(X, dtype= tf.float32)
y = tf.constant(Y, dtype= tf.float32)

x, y
```

```python
input_shape = x[0].shape
output_shape = y[0].shape

input_shape, output_shape
```

```python
plt.scatter(x, y)
```

Steps for modelling with TensorFlow

- Create a model = defining the input and output layers, hidden layers of Deep learning model

- Compiling a model = define loss function (to tell how wrong the model is), optimizer (onforms model how to improve the patterns it is learning), and evaluation metric ( what features can be used to predict model performance)

- Fitting of model = Allowing model to find patterns between x and y

```python
#  finding sequential for the model defining

#  group linear stack of layers into a tf.keras.model by providing training and inference features on this model

```

```python
#  importing sequential model
from tensorflow.keras.models import Sequential

#  set random seed

tf.random.set_seed(42)

#  Creating model using sequential API

model = tf.keras.Sequential( [

     tf.keras.layers.Dense(1)
] )

#  Compile the model ( using mean absolute error - MSE - compares the predicted and observed functions)
model.compile(
              loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.SGD(),
              metrics=["mae"])

#  Fit he model to let it find out patterns/ relationships between x and y

model.fit( tf.expand_dims(x, axis = -1), y, epochs = 5)

```

```python
#  checking x and y

x, y
```

```python
#  checking and making prediction using our model


import numpy as np

pred_y = model.predict(np.array([[17.0]] ))
pred_y
```

```python
pred_y + 11
```

```python
#  Improving the model:

# 1.  Adding the hidden layers, increase no. of neurons, change activation functions of each layers
# 2.  Change optimization function/ learning rate
# 3.  Provide more examples to model to learn and predict
```

```python
# Improving model by increasing the epoch to 100

#  Creating model using sequential API

Imp_model = tf.keras.Sequential( [

     tf.keras.layers.Dense(1)

] )

#  Compile the model ( using mean absolute error - MSE - compares the predicted and observed functions)

Imp_model.compile(
              loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.SGD(),
              metrics=["mae"])

#  Fit he model to let it find out patterns/ relationships between x and y

Imp_model.fit( tf.expand_dims(x, axis = -1), y, epochs = 100)

```

```python
#  checking the data
x, y
```

```python
#  predicting the outputs


import numpy as np

# Convert your list to a 2D NumPy array
input_data = np.array([[17.0]])

# Run the prediction
prediction = Imp_model.predict(input_data)
print(prediction)
```

```python
#  Improving model by increasing hidden layers and neurons count

im_model = tf.keras.Sequential([

          tf.keras.layers.Dense(100, activation="relu"),
          tf.keras.layers.Dense(1)
])

im_model.compile( loss = "mae",
              optimizer= tf.keras.optimizers.SGD(),
               metrics= ["mae"] )

im_model.fit( tf.expand_dims(x, axis = -1), y, epochs = 100)
```

```python
x, y
```

```python
import numpy as np

# Convert your list to a 2D NumPy array
input_data = np.array([[17.0]])

# Run the prediction
prediction = im_model.predict(input_data)
print(prediction)
```

```python
#  Previous model output is closest to the correct / predicted output
#  The model is overfitted, training data is learned very well by model

```

```python
#  Improving model by increasing hidden layers  and neurons count by 50

im1_model = tf.keras.Sequential([

          tf.keras.layers.Dense(50, activation=None),
          tf.keras.layers.Dense(1)
])

# changin g the SGD model with Adam
#  improvimg learning rate by 10 ,  lr is important hyperparamter for a model

im1_model.compile( loss = "mae",
              optimizer= tf.keras.optimizers.Adam(learning_rate= 0.01),
               metrics= ["mae"] )

im1_model.fit( tf.expand_dims(x, axis = -1), y, epochs = 100)
```

```python
x,y
```

```python
import numpy as np

# Convert your list to a 2D NumPy array
input_data = np.array([[17.0]])

# Run the prediction
prediction = im1_model.predict(input_data)
print(prediction)
```

```python
#  Three (3) sets of data in ML

#  Splitting the data into training and testing data

#  Training data = Model learns from the data, 70 - 80 % of original available data
#  Validation set = Model gets tuned on this data, 10 - 15% of available data
#  Testing data = MOdel is evaluated on this data to test the learnings, 10 - 20% of available data
```

```python
#  Length of the data

len(P)
```
