Intro to Tensorflow

```python
import tensorflow as tf

print(tf.__version__)  # version =  2.19.0
```

```python
# creating a constant tensor for a tensor like object with tf.constant()
#  scalar = a single number

scalar = tf.constant(6)
scalar
```

```python

#  checking the number of dimensions of a tensor ( ndim = no. of dimensions)

scalar.ndim
```

```python
#  creating a vector
#  vector = a number with direction( eg. : direction and wind speed)


vector = tf.constant([10, 20])
vector
```

```python
#  dimemsion of vector = 1

vector.ndim
```

```python
#  creating matrix
#  Matrix = 2-D array of numbers

matrix = tf.constant( [[1, 5],
                      [9, 3] ])

matrix
```

```python
matrix.ndim
```

```python

#  creating matrix with float data type

matrix_new = tf.constant( [ [ 9., 2.], [5., 3.], [6., 8.]], dtype = tf.float16)

matrix_new
```

```python
matrix_new.ndim
```

```python
#  creating a tensor
#  Tensor = n dimensional array of numbers ( n can be any numbers )
#  0-D tensor = scalar
#  1-D tensor = vector

tensor = tf.constant( [ [ [3, 6, 1], [2, 9, 2]], [[5, 2, 8], [ 2, 6, 3]], [[9, 1, 4], [1, 2, 1]] ])

tensor
```

```python
tensor.ndim
```

```python
#  creating tensor using tf.variable

changable_tensor = tf.Variable( [10, 6])

unchangeable_tensor = tf.constant( [4, 1])

changable_tensor, unchangeable_tensor
```

```python
#  changing the tensor value using .assign

changable_tensor[0].assign(5)

changable_tensor
```

#  creating random tensors
#  Random tensors = tensors with arbitary size which consists anyrandom numbers
#  Overview of neural network learning ==> initialize random weights, give examples, update representation outputs and repeat with more examples


```python
# creating random tensors of same size
#  seed for reproductivi
random_1 = tf.random.Generator.from_seed(42)
random_1 = random_1.normal( shape=( 3, 2))

random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape =(3,2))

#  checking if both are equal
random_1, random_2, random_1 == random_2
```

Shuffling tensor data

```python
tens = tf.constant([[5, 2], [9, 4], [3, 1] ])

#  suffing the tens tensor based on first dimension ( the pairs remain same only order of pairs channges)
tf.random.shuffle(tens)
```

```python
#  suffles the data every time the cmd is run
tf.random.shuffle(tens)

#  use set_seed to get the same result everytime the command is run
tf.random.set_seed(42)              # global level random seed
tf.random.shuffle(tens, seed=42)    # operation level random seed
```

```python
tf.random.set_seed(25)
tf.random.shuffle(tens, seed=25)
```
