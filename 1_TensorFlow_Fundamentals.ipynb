# Creating random tensors

 Random tensors = tensors with arbitary size which consists any random numbers

Overview of neural network learning ==> initialize random weights, give examples, update representation outputs and repeat with more examples


```python
# creating random tensors of same size
#  seed for reproductivi
random_1 = tf.random.Generator.from_seed(42)
random_1 = random_1.normal( shape=( 3, 2))

random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape =(3,2))

#  checking if both are equal
random_1, random_2, random_1 == random_2
```

Shuffling tensor data

```python
tens = tf.constant([[5, 2], [9, 4], [3, 1] ])

#  suffing the tens tensor based on first dimension ( the pairs remain same only order of pairs channges)
tf.random.shuffle(tens)
```

```python
#  suffles the data every time the cmd is run
tf.random.shuffle(tens)

#  use set_seed to get the same result everytime the command is run
tf.random.set_seed(42)              # global level random seed
tf.random.shuffle(tens, seed=42)    # operation level random seed
```

```python
tf.random.set_seed(25)
tf.random.shuffle(tens, seed=25)
```

# Tensor **attributes** to fetch information from tensors

Shape = Length/ number of elements of each dimensions of a tensor ( tensor.shape)

Rank = Number of tensor dimensions( scalar = 0, vector = 1, matrix = 2, tensor = n ) ( tensor.ndim)

Axis / dimension = Sprcific dimension of a tensor ( tensor[0], tensor[:, 1], ...

size = Total number of items in tensor ( tf.size( tensor) )

```python
#  creating tensor of rank 4 ( 4 D tensor )

rank_4_tens = tf.ones( shape =( 2, 4, 2, 5))

rank_4_tens
```

```python
rank_4_tens[0]
```

```python
rank_4_tens.shape, rank_4_tens.ndim, tf.size( rank_4_tens)
```

```python

#  different attributes of tensor

print( " Datatype of every elements : ", rank_4_tens.dtype)

print( " Number of dimensions(rank) : ", rank_4_tens.ndim)

print( " Shape of tensor : " , rank_4_tens.shape)
print( " Elements along 0 axis : ", rank_4_tens.shape[0])
print(" Elements along last axis : ", rank_4_tens.shape[-1])
print( " Total number of elements in tensor : ", tf.size( rank_4_tens))

print( " Total number of elements in tensor : ", tf.size( rank_4_tens).numpy())  # single count of elements
```

```python
#  creating a rank 2 tensor ( matrix)

rank2_tens = tf.constant([[5, 2], [9, 4]])

rank2_tens.shape, rank2_tens.ndim
```

```python
#  Extracting last item from each row

rank2_tens[:-1]
```

```python
#  Adding extra dimension to rank2 tensor, keeping old data same
#  newaxis adds a new dimension to the existing tensor at last axis

rank3_tens = rank2_tens[..., tf.newaxis]
rank3_tens
```

```python
#  tf.expand_dim() , alternate to tf.newaxis() to expand final axis(-1) in existing tensor

tf.expand_dims(rank2_tens, axis = -1)
```

```python
#  expanding the 0th axis

tf.expand_dims(rank2_tens, axis = 0)
```

```python
#  expanding the 1st axis

tf.expand_dims(rank2_tens, axis = 1)
```
